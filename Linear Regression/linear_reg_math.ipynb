```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
weight = np.linspace(50, 100, 100)
height = 4*weight + np.random.random(100)*20
```

```python
plt.plot(weight, height, '.')
plt.xlabel('Weight')
plt.ylabel('Height')
```

```python
W = np.vstack([np.ones_like(weight), weight]).T
W
```

$$\beta = (W^TW)^{-1}W^TH$$

```python
beta = np.matmul(np.matmul(np.linalg.inv(np.matmul(W.T, W)), W.T), height)
beta
```

```python
plt.plot(weight, height, '.')
plt.plot(weight, beta[1]*weight + beta[0])
plt.xlabel('Weight')
plt.ylabel('Height')
```

### `np.linalg.lstsq`

```python
W = np.vstack([np.ones_like(weight), weight]).T
W
```

```python
beta = np.linalg.lstsq(W, height, rcond=None)[0]
beta
```

```python
plt.plot(weight, height, '.')
plt.plot(weight, beta[1]*weight + beta[0])
plt.xlabel('Weight')
plt.ylabel('Height')
```
