```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Datasets/creditcard.csv')
df.head(10)
```

```python
df.iloc[:,:-1].describe()
```

```python
for c in df.columns[:-1]:
    df.boxplot(c,by='Class',figsize=(7,4),fontsize=14)
    plt.title("{}\n".format(c),fontsize=16)
    plt.xlabel("Variable")
```

## PCA

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = df.drop(['Time','Class'],axis = 1)
y = df['Class']
```

```python
X = scaler.fit_transform(X)
```

```python
dfx = pd.DataFrame(data=X,columns=df.columns[1:-1])
dfx.head(10)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=len(df.columns[1:-1]))
```

```python
pca.fit(dfx)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=[i+1 for i in range(len(pca.explained_variance_ratio_))],y=pca.explained_variance_ratio_,s=200,alpha=0.8,edgecolors='k')
plt.grid(True)
plt.title("Explained variance ratio of the \nfitted principal component vector\n",fontsize=25)
plt.xlabel("Principal components",fontsize=15)
plt.yticks(fontsize=15)
plt.xticks([i+1 for i in range(len(pca.explained_variance_ratio_))],fontsize=15)
plt.ylabel("Explained variance ratio",fontsize=15)
plt.show()
```

Hence we conclude PCA cannot be used for this scenario
