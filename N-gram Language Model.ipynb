```python
f = open('../Deep Learning/Pytorch-Tutorial/Natural Language Processing/.data/multi30k/train.en', 'r')
lines2 = f.readlines()
```

```python
lines = [line.split() for line in lines2]
lines
```

```python
n_grams = {}
```

```python
for sentence in lines:
    words = []
    for word in sentence:
        if word[0].isalpha():
            if word[-1] == '.':
                words.append(word[:-1])
                words.append('.') # Separate full stop from word
            else:
                words.append(word)
    for ix in range(len(words) - 1): # Except last word ('.')
        try:
            n_grams[words[ix]].append(words[ix+1])
        except KeyError as _:
            n_grams[words[ix]] = []
            n_grams[words[ix]].append(words[ix+1])
n_grams # All bigrams, with what words follow a certain word
```

```python
import random
```

```python
def generate_sentence(nb = 7):
    words = []
    next_word = random.choice(list(n_grams.keys())) # Random key
    words.append(next_word) # The first word
    while len(words) < nb:
        next_word = random.choice(n_grams[next_word])
        words.append(next_word)
    return " ".join(words) # Join list with space in between
```

```python
generate_sentence(7)
```
